```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
def linContrastStretching(img, a, b):
    return np.uint8(np.round(a+(img-np.min(img))*((b-a)/(np.max(img) - np.min(img)))))
```

```python
img = cv.imread('./DIP_2019_A1/fg.jpg')
```

```python
out = linContrastStretching(copy.deepcopy(img), 100, 200)
```

```python
_, images = plt.subplots(1,2, figsize = (20,20))
images[0].imshow(img, cmap='gray', vmin = 0, vmax = 255)
images[0].set_title('original')
images[1].imshow(out, cmap='gray', vmin = 0, vmax = 255)
images[1].set_title('contrast streched')
plt.show()
```

```python
def colorBar(img, k):
    img = img.reshape((-1,img.shape[-1]))
    colors, counts = np.unique(img, return_counts=True, axis=0)
    sort_ind = np.argsort(counts)[::-1]
    return np.reshape((colors[sort_ind[0:k]].T)[0], (-1, k)) 
    
```

```python
outp = colorBar(img,10).T
print(outp)

_, cb = plt.subplots()
cb.set_xticks(np.arange(-.5, outp.shape[1], 1));
cb.set_yticks(np.arange(-.5, outp.shape[0], 1));
cb.set_xticklabels(np.arange(0, outp.shape[1]+1, 1))
cb.set_yticklabels(np.arange(1, outp.shape[0]+1, 1))
cb.imshow(outp, cmap = 'gray', vmin = 0, vmax = 255 )

cb.grid(color='w', linewidth=2)
cb.set_frame_on(False)
plt.show()
```
